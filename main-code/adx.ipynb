```python
import talib as ta
import pandas as pd
import numpy as np
```

```python
# read the data from the CSV files
train_data = pd.read_csv('train_data_50.csv') 

# change date into datetime objects
train_data['Date'] = pd.to_datetime(train_data['Date'])

# set indexes 
train_data.set_index(["Ticker", "Date"], inplace=True)

# sort data by tickers
tickers = sorted(train_data.index.get_level_values('Ticker').unique())

# array of all stocks and close prices
open_prices = []
high_prices = []
low_prices = []
```

```python
#CLOSE
for ticker in tickers:
    stock_close_data = train_data.loc[ticker]["Open"]
    open_prices.append(stock_close_data.values)

open_prices = np.stack(open_prices)
print(open_prices.shape)
print(open_prices)

# array for final return - trades
trades = np.zeros_like(open_prices)
trades
```

```python
#LOW

for ticker in tickers:
    stock_low_data = train_data.loc[ticker]["Low"]
    low_prices.append(stock_close_data.values)

low_prices = np.stack(low_prices)
print(low_prices.shape)
print(low_prices)

```

```python
#HIGH

for ticker in tickers:
    stock_high_data = train_data.loc[ticker]["High"]
    high_prices.append(stock_high_data.values)

high_prices = np.stack(high_prices)
print(high_prices.shape)
print(high_prices)

# array for final return - trades
```

```python
# implement A/D Index logic

for stock in range(len(high_prices)):
    talib_adx = ta.ADX(stock-stock_high_data, stock-stock_low_data, stock-stock_close_data, timeperiod = 15)
    talib_dmi = ta.DX(stock-stock_high_data, stock-stock_low_data, stock-stock_close_data, timeperiod = 15)
    
    for day in range(1,len(high_prices[0]) - 1):
        if talib_adx[day] > 25 and high_prices[stock] > high_prices[stock-1] :
            trades[stock][day+1] = -1 
        elif talib_adx[day] > 25 and high_prices[stock] < high_prices[stock-1] :
            trades[stock][day+1] = 1 
        else :
            trades[stock][day+1] = 0
        
print(trades)
```
